```python
df= {'A':'American Home Food Products', 'G':'General Mills', 'K':'Kelloggs','N':'Nabisco',
     'P':'Post', 'Q':'QuakerOats', 'R':'Ralston Purina'}
cereals1 = cereals.replace({'mfr':df})
cereals1['mfr']
```

```python
cereals.sort_values(by='rating',ascending=False).head().reset_index(drop=True)
```

```python
df1=cereals1.groupby('mfr')['rating'].mean()
df1.plot.bar()
```

```python
count = cereals1['mfr'].value_counts().reset_index()
plt.pie(count['mfr'],labels=count['index'])
```
