```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
house = pd.read_csv("housing_price.csv")
house
```

```python
house.isnull().sum()
```

```python
new_house = house.dropna()
new_house
```

```python
x = new_house[["longitude","latitude","housing_median_age","total_rooms","total_bedrooms","population","households","median_income"]].values
print("independent data",x)
y = new_house['median_house_value'].values
print("dependent data",y)
```

```python
x =new_house.drop("median_house_value",axis = 1)
print("independent data",x)
y = new_house['median_house_value'].values
print("dependent data",y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
print("X_train and x_test dataset shape",(X_train.shape),(X_test.shape))
print("y_train and y_test dataset shape",(y_train.shape),(y_test.shape))
```

```python
from sklearn.metrics import r2_score
X =new_house.drop("median_house_value",axis = 1)
y = new_house['median_house_value'].values
reg = LinearRegression()
reg.fit(X, y)
reg.score(X,y)
print("r value is",reg.score(X,y))
print("reg.coefficient is",reg.coef_)
print("intercept is",reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head(9)
```
