```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('netflix_titles.csv')
df
```

```python
df1=df['type'].value_counts().sort_values()
plt.pie(df1,labels = ['Movie','TV Show '],explode = [0.1,0.1],autopct="%1.2f%%",colors = ['blue','red'],startangle =90)
plt.title('Type of Netflix content')
        
```
