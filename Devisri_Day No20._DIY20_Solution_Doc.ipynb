```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import iqr
employ_data = pd.read_csv("HRDataset_v14.csv")
employ_data
```

```python
employ_data1 = employ_data.groupby(['Department'])
employ_data1.first()
```

```python
employ_salary= employ_data.groupby(['Department'],as_index= False).Salary.mean()
employ_salary
```

```python
employ_salary1= employ_data.groupby(['Department'],as_index= False).Salary.median()
employ_salary1
```

```python
employ_salary= employ_data['Department'].mode()
display("This department has highest number of employees")
employ_salary
```

```python
employ_salary= employ_data.groupby(['Department']).Salary.median().std()
display("department wise standard deviation of salary")
employ_salary
```

```python
employ_salary= employ_data.groupby(['Department']).Salary.median().var()
display("department wise variance of salary")
employ_salary
```

```python
from scipy.stats import iqr
employ = iqr(employ_salary1['Salary'],axis=0)
display("Interquartile range of salary")
employ
```
