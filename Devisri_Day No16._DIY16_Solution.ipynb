```python
import pandas as pd
df1 = pd.read_csv("SampleSuperstore.csv")
df1
```

```python
plt.bar('Category', 'Sales', data = df1)
plt.xlabel("Category")
plt.ylabel("Sales")
  
plt.show()
```

```python
df2 = df1.groupby('State')['Sales'].count().sort_values(ascending = False).reset_index()
plt.figure(figsize=(20,10))
plt.bar('State','Sales',data =df2)
plt.xlabel('State')
plt.ylabel('Sales')
plt.xticks(rotation =90)
plt.title('TotaL SALES value of every state')
```
