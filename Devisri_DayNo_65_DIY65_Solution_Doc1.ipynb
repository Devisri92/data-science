```python
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
df1=pd.read_csv("voice.csv")
df1
```

```python
df2=df1.rename(columns={"label":"Gender_Identified"})
df2
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df2['Gender_Identified']= label_encoder.fit_transform(df2['Gender_Identified'])
df2['Gender_Identified'].unique()
df2
```

```python
x = df2.drop(columns = "Gender_Identified")
y = df2.Gender_Identified
```
