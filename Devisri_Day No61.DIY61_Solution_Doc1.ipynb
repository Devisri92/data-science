```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.model_selection import train_test_split
from warnings import filterwarnings
filterwarnings(action='ignore')
df1 = pd.read_csv("winequality-white.csv",sep=";")
df1
```

```python
df1.isnull().sum()
```

```python
df1.info()
```

```python
x = df1.drop(columns ="alcohol")
y = df1.alcohol
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 100)
print("shape of train data is :",x_train.shape,y_train.shape)
print("shape of test data is :",x_test.shape,y_test.shape)
```

```python
data = {'Model Name': [], 'MAPE': [],'RMSE':[],'RMSLE':[]}
df2 = pd.DataFrame(data)
df2
```

```python
from sklearn.linear_model import LinearRegression
model_ler = LinearRegression().fit(x_train,y_train)
pred = model_ler.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df3 = {'Model Name': 'Linear regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df2 = df2.append(df3, ignore_index = True)
df2
```

```python
# svm 
from  sklearn.svm import SVR
model_svr = SVR().fit(x_train,y_train)
pred = model_svr.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df3 = {'Model Name': 'SVM regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df2 = df2.append(df3, ignore_index = True)
df2
```

```python
#ridge regression
from sklearn.linear_model import Ridge
ridgeReg = Ridge(alpha=0.0005, normalize=True)
ridgeReg.fit(x_train,y_train)
pred = ridgeReg.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df3 = {'Model Name': 'Ridge regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df2 = df2.append(df3, ignore_index = True)
df2
```

```python
from sklearn.linear_model import Lasso
lassoReg = Lasso(alpha=0.0005, normalize=True)
lassoReg.fit(x_train,y_train)
pred = lassoReg.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df3 = {'Model Name': 'Lasso regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df2= df2.append(df3, ignore_index = True)
df2
```

```python
from sklearn.tree import DecisionTreeRegressor

dtreg= DecisionTreeRegressor(max_depth=5)
dtreg.fit(x_train,y_train)
pred = dtreg.predict(x_test)

# RMSE
rmse = np.sqrt(np.mean((y_test - pred)**2))

# MAPE
mape = (np.mean(np.abs(((y_test - pred)/y_test))))*100

#RMSLE
rmsle = np.sqrt(np.square(np.log(pred + 1) - np.log(y_test + 1)).mean())

df3 = {'Model Name': 'Decision Tree regression', 'MAPE':mape,'RMSE':rmse,'RMSLE':rmsle}
df2 = df2.append(df3, ignore_index = True)
df2
```
