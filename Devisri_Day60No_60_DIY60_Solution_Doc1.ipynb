```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
df1=pd.read_csv('titanic.csv')
df1
```

```python
df1.isnull().sum()
```

```python
df2=df1.drop(['PassengerId', 'Age','Cabin'], axis=1)
df2
```

```python
df3 = df2.dropna(how='any')
df3
```

```python
df3.isnull().sum()
```

```python
X = df3.drop(["Name","Sex","Embarked","Ticket"],axis=1)
y= df3["Survived"]
```

```python
X
```

```python
y
```

```python
kf=KFold(n_splits=5)
kf.get_n_splits(X,y)
```

```python
clf=DecisionTreeClassifier()
for train_ind, test_ind in kf.split(X,y):
    clf.fit(X.iloc[train_ind,:],y.iloc[train_ind])
    pred=clf.predict(X.iloc[test_ind,:])
    print("Accuracy For Each fold of data are")
    print(round(accuracy_score(y.iloc[test_ind],pred),3))
   
```

```python
cv =cross_val_score(DecisionTreeClassifier(),X,y,cv=5,scoring='accuracy')
print("Accuracy of all the folds after the cross validation are",cv)
```
